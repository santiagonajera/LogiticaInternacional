```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Paso 1: Descargar el archivo Excel desde el enlace
url = "https://github.com/santiagonajera/LogiticaInternacional/raw/refs/heads/main/logistica_internacional_ecuador.xlsx"
df = pd.read_excel(url)

# Paso 2: Inspeccionar las primeras filas del DataFrame
print(df.head())

# Paso 3: Procesar la columna 'percepcion_calidad_cliente'
# Convertir los valores de texto (con coma como separador decimal) a números flotantes
df['percepcion_calidad_cliente'] = df['percepcion_calidad_cliente'].str.replace(',', '.').astype(float)

# Paso 4: Generar el histograma
plt.figure(figsize=(10, 6))
sns.histplot(
    data=df,
    x='percepcion_calidad_cliente',
    bins=10,  # Número de intervalos en el histograma
    kde=True,  # Agregar una curva de densidad
    color='skyblue'
)

# Personalizar el gráfico
plt.title('Histograma de Percepción de Calidad del Cliente', fontsize=16)
plt.xlabel('Percepción de Calidad', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Paso 1: Descargar el archivo Excel desde el enlace
url = "https://github.com/santiagonajera/LogiticaInternacional/raw/refs/heads/main/logistica_internacional_ecuador.xlsx"
df = pd.read_excel(url)

# Paso 2: Verificar y procesar la columna 'percepcion_calidad_cliente'
if df['percepcion_calidad_cliente'].dtype == 'object':  # Si es de tipo string
    # Reemplazar comas por puntos y convertir a float
    df['percepcion_calidad_cliente'] = df['percepcion_calidad_cliente'].str.replace(',', '.').astype(float)
elif df['percepcion_calidad_cliente'].dtype in ['float64', 'int64']:  # Si ya es numérico
    print("La columna ya está en formato numérico.")
else:
    # Manejar otros casos (por ejemplo, valores nulos o mixtos)
    df['percepcion_calidad_cliente'] = pd.to_numeric(df['percepcion_calidad_cliente'], errors='coerce')

# Paso 3: Eliminar valores nulos (si existen)
df = df.dropna(subset=['percepcion_calidad_cliente'])

# Paso 4: Generar histogramas separados por 'direccion_comercio'
# Crear una figura con subgráficos para cada categoría de 'direccion_comercio'
g = sns.FacetGrid(
    df,
    col="direccion_comercio",  # Separar por la columna 'direccion_comercio'
    height=5,  # Altura de cada subgráfico
    aspect=1.2,  # Proporción ancho/alto
    sharex=True,  # Compartir el eje X entre los subgráficos
    sharey=False  # No compartir el eje Y
)

# Añadir histogramas a cada subgráfico
g.map_dataframe(
    sns.histplot,
    x="percepcion_calidad_cliente",
    bins=5,  # Número de intervalos en el histograma
    kde=False,  # Sin curva de densidad
    stat="count",  # Mostrar frecuencia absoluta
    color=None  # Usar paleta de colores
)

# Asignar colores diferentes a las barras de cada histograma
palette = sns.color_palette("husl", len(df["direccion_comercio"].unique()))
for ax, color in zip(g.axes.flat, palette):
    for patch in ax.patches:
        patch.set_facecolor(color)

# Personalizar los títulos y etiquetas
g.set_titles("{col_name}", fontsize=14)  # Título de cada subgráfico
g.set_axis_labels("Percepción de Calidad", "Frecuencia", fontsize=12)  # Etiquetas de los ejes
g.tight_layout()

# Mostrar el gráfico
plt.show()
```
