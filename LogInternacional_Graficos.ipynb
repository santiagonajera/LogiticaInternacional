```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Paso 1: Descargar el archivo Excel desde el enlace
url = "https://github.com/santiagonajera/LogiticaInternacional/raw/refs/heads/main/logistica_internacional_ecuador.xlsx"
df = pd.read_excel(url)

# Paso 2: Verificar las columnas relevantes
print("Valores únicos en 'pais_destino':", df['pais_destino'].unique())
print("Valores únicos en 'pais_origen':", df['pais_origen'].unique())

# Paso 3: Crear una columna categórica para contar la frecuencia de 'pais_origen'
# Contamos cuántas veces aparece cada país de origen para cada país de destino
df_counts = df.groupby(['pais_destino', 'pais_origen']).size().reset_index(name='frecuencia')

# Paso 4: Generar histogramas separados por 'pais_destino'
# Crear una figura con subgráficos para cada categoría de 'pais_destino'
g = sns.FacetGrid(
    df_counts,
    col="pais_destino",  # Separar por la columna 'pais_destino'
    col_wrap=3,  # Máximo 3 subgráficos por fila
    height=4,  # Altura de cada subgráfico
    aspect=1.2,  # Proporción ancho/alto
    sharex=False,  # No compartir el eje X entre los subgráficos
    sharey=False  # No compartir el eje Y
)

# Añadir histogramas a cada subgráfico
g.map_dataframe(
    sns.histplot,
    x="pais_origen",
    weights="frecuencia",  # Usar la frecuencia como peso
    bins=5,  # Número de intervalos en el histograma
    kde=False,  # Sin curva de densidad
    stat="count"  # Mostrar frecuencia absoluta
)

# Asignar colores diferentes a las barras de cada histograma
palette = sns.color_palette("husl", len(df["pais_destino"].unique()))
for ax, color in zip(g.axes.flat, palette):
    for patch in ax.patches:
        patch.set_facecolor(color)

# Personalizar los títulos y etiquetas
g.set_titles("{col_name}", fontsize=12)  # Título de cada subgráfico
g.set_axis_labels("País de Origen", "Frecuencia", fontsize=10)  # Etiquetas de los ejes
g.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Paso 1: Descargar el archivo Excel desde el enlace
url = "https://github.com/santiagonajera/LogiticaInternacional/raw/refs/heads/main/logistica_internacional_ecuador.xlsx"
df = pd.read_excel(url)

# Paso 2: Convertir 'fecha_entrega_real' a formato datetime
df['fecha_entrega_real'] = pd.to_datetime(df['fecha_entrega_real'], errors='coerce')

# Paso 3: Crear una columna con el formato 'mes-año'
df['mes_año'] = df['fecha_entrega_real'].dt.to_period('M').astype(str)

# Paso 4: Agrupar por 'mes-año' y sumar las cantidades pedidas y entregadas
agrupado = df.groupby('mes_año').agg({
    'cantidad_pedida': 'sum',
    'cantidad_entregada': 'sum'
}).reset_index()

# Paso 5: Ordenar los datos por fecha (mes-año)
agrupado['mes_año'] = pd.to_datetime(agrupado['mes_año'], format='%Y-%m')
agrupado = agrupado.sort_values(by='mes_año')

# Paso 6: Generar el gráfico de áreas superpuesto
plt.figure(figsize=(12, 6))

# Área para cantidad pedida
plt.fill_between(
    agrupado['mes_año'],
    agrupado['cantidad_pedida'],
    label='Cantidad Pedida',
    color='red',
    alpha=0.5
)

# Área para cantidad entregada
plt.fill_between(
    agrupado['mes_año'],
    agrupado['cantidad_entregada'],
    label='Cantidad Entregada',
    color='skyblue',
    alpha=0.6
)

# Personalizar el gráfico
plt.title('Comparación de Cantidad Pedida vs. Cantidad Entregada por Mes-Año', fontsize=16)
plt.xlabel('Mes-Año', fontsize=12)
plt.ylabel('Cantidad', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Formatear el eje X para mostrar solo el mes y año
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.MonthLocator(interval=1))
plt.xticks(rotation=45)

# Ajustar la escala del eje Y para eliminar espacios en blanco
plt.ylim(bottom=0)  # Fuerza el eje Y a comenzar en 0
plt.margins(x=0)    # Elimina márgenes en el eje X

# Añadir leyenda
plt.legend(loc='upper left', fontsize=10)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Paso 1: Descargar el archivo Excel desde el enlace
url = "https://github.com/santiagonajera/LogiticaInternacional/raw/refs/heads/main/logistica_internacional_ecuador.xlsx"
df = pd.read_excel(url)

# Función para calcular el OTIF
def calcular_otif(df):
    # Calcular cumplimiento de "In Full" (IF)
    df['cumple_IF'] = (df['cantidad_entregada'] >= df['cantidad_pedida']).astype(int)
    porcentaje_IF = df['cumple_IF'].mean() * 100  # Porcentaje de cumplimiento IF

    # Calcular cumplimiento de "On Time" (OT)
    df['cumple_OT'] = (df['tiempo_total_real_dias'] <= df['tiempo_total_ofertado_dias']).astype(int)
    porcentaje_OT = df['cumple_OT'].mean() * 100  # Porcentaje de cumplimiento OT

    # Calcular OTIF
    OTIF = (porcentaje_IF / 100) * (porcentaje_OT / 100) * 100  # Porcentaje final del OTIF
    return OTIF

# Paso 2: Separar los datos por dirección de comercio
df_importaciones = df[df['direccion_comercio'] == 'Importación']
df_exportaciones = df[df['direccion_comercio'] == 'Exportación']

# Paso 3: Calcular el OTIF para importaciones y exportaciones
otif_importaciones = calcular_otif(df_importaciones)
otif_exportaciones = calcular_otif(df_exportaciones)

# Paso 4: Definir las zonas del gráfico de viñeta
meta = 81.9  # Meta objetivo
zona_azul = meta
zona_tomate = 70.0
zona_roja = 0.0

# Colores metalizados
colores = ['#C0C0C0', '#A9A9A9', '#808080']  # Plateado, Gris claro, Gris oscuro

# Función para generar el gráfico de viñeta
def grafico_viñeta(otif, titulo):
    fig, ax = plt.subplots(figsize=(10, 1))  # Gráfico más alargado

    # Zonas de fondo
    ax.barh([0], [zona_azul], color=colores[0], edgecolor='black', height=0.4, label='Zona Azul (≥81.9%)')
    ax.barh([0], [zona_tomate], color=colores[1], edgecolor='black', height=0.4, label='Zona Tomate (70%-81.9%)')
    ax.barh([0], [100], color=colores[2], edgecolor='black', height=0.4, label='Zona Roja (<70%)')  # Zona roja completa

    # Valor del OTIF
    ax.scatter(otif, 0, color='black', s=100, label=f'OTIF: {otif:.1f}%')  # Punto que muestra el valor del OTIF
    ax.plot([otif, otif], [0 - 0.2, 0 + 0.2], color='black', linewidth=2)  # Línea vertical

    # Personalización del gráfico
    ax.set_xlim(0, 100)  # Rango del eje X
    ax.set_ylim(-0.5, 0.5)  # Rango del eje Y
    ax.set_yticks([])  # Eliminar ticks del eje Y
    ax.set_xlabel('Porcentaje (%)', fontsize=10)
    ax.set_title(titulo, fontsize=12, pad=20)

    # Añadir leyenda
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles[::-1], labels[::-1], loc='upper center', bbox_to_anchor=(0.5, -0.5), ncol=3, fontsize=8)

    # Mostrar el gráfico
    plt.tight_layout()
    plt.show()

# Paso 5: Generar los gráficos separados
grafico_viñeta(otif_importaciones, 'Desempeño OTIF - Importaciones')
grafico_viñeta(otif_exportaciones, 'Desempeño OTIF - Exportaciones')
```
